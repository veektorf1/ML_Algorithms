```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.001,1,200,endpoint=True)
y1 = -np.log2(x)*x
y2 = -np.log2(1-x)*(1-x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()

plt.plot(x,y1+y2) # Sum of two classes
plt.xlabel('probability of occurrence')
plt.show()
```

```python

def entropy(y):
    """
        This function calculates entropy for given input y
        @param y: list of labels 
        @return: entropy value
    """
    unique,count = np.unique(y,return_counts=True)
    # Entopy formula: H(Y) = -p(y)log2(p(y))

    prob = count/len(y) # stores array of probabilities for each class
    log_prob = np.log2(prob) # stores array of log2 probabilities for each class
    entropy = -np.sum(prob*log_prob) 
    return entropy

data = np.zeros(100)
print(entropy(np.hstack((data,[1])))) # 0.08
print(entropy(np.hstack((data,[1,1])))) # 0.19
"""
Entropy properties:
    1. Entropy is maximum (1) when all classes are equally likely
    2. Entropy is minimum (0) when only one class is present
    3. Entropy penalizes impurity strongly. 
        - Above example with 100 zeros and 1 one has entropy of 0.08
        - Above example with 100 zeros and 2 ones has entropy of 0.19
"""
```
